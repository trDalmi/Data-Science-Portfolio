```python
# Initialize geolocator
geolocator = Nominatim(user_agent="city_finder")

def get_city_from_pincode(pincode):
    try:
        # Geocode the pincode
        location = geolocator.geocode(str(pincode),language="en")  # Specify country for better accuracy

        if location:
            return location.address.split(',')[-2].strip() # Extract city name from the address
        else:
            return "City not found"

    except Exception as e:
        return "City not found"
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd
import time

# Configure geocoder with proper settings
geolocator = Nominatim(
    user_agent="my-geo-app-contact@example.com",
    timeout=10  # Increased timeout duration
)

# Create rate-limited geocoder with retries
geocode = RateLimiter(
    geolocator.geocode,
    min_delay_seconds=1.2,
    max_retries=2,
    error_wait_seconds=5
)

def get_country(row):
    city = str(row['City']).strip()

    try:
        # Perform city search
        location = geocode(
            {"city": city},
            exactly_one=True,
            language="en",
            addressdetails=True
        )

        if location and location.raw.get('address'):
            country = location.raw['address'].get('country')
            if country:
                return country

        return "Not found"

    except Exception as e:
        print(f"Error processing {city}: {str(e)}")
        time.sleep(2)  # Additional delay on error
        return "Error"
```
