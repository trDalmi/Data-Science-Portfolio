```python
df = df.dropna()
```

```python
df = df.reset_index(drop=True)
df
```

```python
# Convert the 'date_added' column to datetime objects
df['Release Date'] = pd.to_datetime(df['Release Date'])

# Extract the month and year
df['month'] = df['Release Date'].dt.month
df['year'] = df['Release Date'].dt.year

# Display the updated dataframe (optional)
df
```

```python
df['Hours Viewed'] = df['Hours Viewed'].str.replace(',', '')
df['Hours Viewed'] = df['Hours Viewed'].astype(float)
df['Hours Viewed'].min()
```

```python
df['Hours Viewed'] = df['Hours Viewed']/df['Hours Viewed'].min()
df['Hours Viewed'].min()
```

```python
# Calculate the frequency of each day of the week
df['Day of Week'].value_counts()
```

For this we can see that most of the content is released on saturday and the followed by friday ,wednesday, thursday, tuesday and then mon and least sun.

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and the column with days of the week is 'Day of Week'
day_counts = df['Day of Week'].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(day_counts.index, day_counts.values)
plt.xlabel("Day of the Week")
plt.ylabel("Number of Releases")
plt.title("Content Release Frequency by Day")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
show_df = df[df['Content Type'] == 'Show']
day_counts_show =  show_df['Day of Week'].value_counts()
movies_df = df[df['Content Type'] == 'Movie']
day_counts_movie =  movies_df['Day of Week'].value_counts()
```

```python
plt.figure(figsize=(6, 6))

# Define the width of each bar
bar_width = 0.25

# Set the positions of the bars on the x-axis
r1 = range(len(day_counts))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]


# Create the bar plots
plt.bar(r1, day_counts.values, color='skyblue', width=bar_width, label='All Content')
plt.bar(r2, day_counts_show.values, color='orange', width=bar_width, label='Shows')
plt.bar(r3, day_counts_movie.values, color='green', width=bar_width, label='Movies')


# Add labels, title, and legend
plt.xlabel("Day of the Week")
plt.ylabel("Number of Releases")
plt.title("Content Release Frequency by Day and Content Type")
plt.xticks([r + bar_width for r in range(len(day_counts))], day_counts.index, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

From above we can say that most of thre realse we focus on friday as of its the start of the weekend, so as to drive more watching the Netflix targets Friday in particluar. Followed by wed, thus, tue, mon, sat and least sun.
And other than that we can see the the Show are more being more prominently release reather thant he movies so that they could hook the viewers up and so that the latch on and try to catch the next episode weekly at the same time. same pattern is followe accros the whole week just on tues the movies are released slighlty more than shows. Coz obvi it mid week and people dont have enough time to binge watch something, they must just prefer something short and content.

```python
#counting the release per months
Release_count_month = df['month'].value_counts()

plt.figure(figsize=(6, 6))
plt.bar(Release_count_month.index, Release_count_month.values)
plt.xlabel("Month")
plt.ylabel("Number of Releases")
plt.title("Content Release Frequency by Month")
```

From the above bar plot we can see that for the starting month the release are lesser and it keeps increasing towards the end of the year, as the holiday season approaches.

```python
# counting different years present in data
Year_count_of_release = df['year'].value_counts()

plt.figure(figsize=(6, 6))
plt.bar(Year_count_of_release.index, Year_count_of_release.values)
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.title("Content Release Frequency by Year")
```

From the above graph we can see that the yearly realse before 2014 was not considerablt large on the netflix platform but after 2014 it incresed with a very large slope and in 2020 and 2021 almost has similar number of relase with a slight increase but it increased considerable in 2022 and but then in 2023 it decreased up to apprx 800 release that is less than from back in 2018, It could be because of the change in the preference in tht type of content they wanna put on the platform to more in quality rather than quanity.

```python
#i want to find out the effect of the day of release on the hourly viewed.
#so for this i could summ up of the hourly viewed for all the weekdays seperatelt and then plot it agisnt the weekdayas.this

sat_hr = df[df['Day of Week'] == 'Sat']
sat_hr_view = sat_hr['Hours Viewed'].sum()
sun_hr = df[df['Day of Week'] == 'Sun']
sun_hr_view = sun_hr['Hours Viewed'].sum()
mon_hr = df[df['Day of Week'] == 'Mon']
mon_hr_view = mon_hr['Hours Viewed'].sum()
tue_hr = df[df['Day of Week'] == 'Tue']
tue_hr_view = tue_hr['Hours Viewed'].sum()
wed_hr = df[df['Day of Week'] == 'Wed']
wed_hr_view = wed_hr['Hours Viewed'].sum()
thu_hr = df[df['Day of Week'] == 'Thu']
thu_hr_view = thu_hr['Hours Viewed'].sum()
fri_hr = df[df['Day of Week'] == 'Fri']
fri_hr_view = fri_hr['Hours Viewed'].sum()

hrs_lst = [sun_hr_view, mon_hr_view, tue_hr_view, wed_hr_view, thu_hr_view, fri_hr_view,sat_hr_view]
days = ['sun','mon','tue','wed','thu','fri','sat']

plt.figure(figsize=(6,6))
plt.bar(days,hrs_lst)
plt.xlabel('Days')
plt.ylabel('Hours Viewed in Million')
plt.title('Effect of Day of Release on Hourly Viewed')
```

From the above data we can say the releases on the friday brought the most hours viewed approx 380000 Million hrs followe by thu,wed and tue sat, mon and least on sun.

```python
# ok same thing but i wanna know the effect of the months on the views
jan_hr = df[df['month'] == 1]
jan_hr_view = jan_hr['Hours Viewed'].sum()
feb_hr = df[df['month'] == 2]
feb_hr_view = feb_hr['Hours Viewed'].sum()
mar_hr = df[df['month'] == 3]
mar_hr_view = mar_hr['Hours Viewed'].sum()
apr_hr = df[df['month'] == 4]
apr_hr_view = apr_hr['Hours Viewed'].sum()
may_hr = df[df['month'] == 5]
may_hr_view = may_hr['Hours Viewed'].sum()
jun_hr = df[df['month'] == 6]
jun_hr_view = jun_hr['Hours Viewed'].sum()
jul_hr = df[df['month'] == 7]
jul_hr_view = jul_hr['Hours Viewed'].sum()
aug_hr = df[df['month'] == 8]
aug_hr_view = aug_hr['Hours Viewed'].sum()
sep_hr = df[df['month'] == 9]
sep_hr_view = sep_hr['Hours Viewed'].sum()
oct_hr = df[df['month'] == 10]
oct_hr_view = oct_hr['Hours Viewed'].sum()
nov_hr = df[df['month'] == 11]
nov_hr_view = nov_hr['Hours Viewed'].sum()
dec_hr = df[df['month'] == 12]
dec_hr_view = dec_hr['Hours Viewed'].sum()

month_hrs_lst = [jan_hr_view, feb_hr_view, mar_hr_view, apr_hr_view, may_hr_view, jun_hr_view,jul_hr_view,aug_hr_view,sep_hr_view,oct_hr_view,nov_hr_view,dec_hr_view]
months = ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize=(6,6))
plt.bar(months,month_hrs_lst)
plt.xlabel('Months')
plt.ylabel('Hours Viewed in Million')
plt.title('Effect of Month of Release on Hourly Viewed')
```

From the above graph we can see the the dec and june had the highest hrlys views compare to the other months. so we can say the the holiday seasons of schools, unis and festival has a considerable effect on the views.
